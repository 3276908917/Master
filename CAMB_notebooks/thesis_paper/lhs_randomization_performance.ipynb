```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wall_times = np.load("../../wall_times.npy")
fn_calls = np.load("../../fn_calls.npy")
bests = np.load("../../bests.npy")
```

```python
fig = plt.figure()
ax = fig.gca()

ax.set_title("$s^*_\mathrm{best}$ versus Wall Time", fontsize=24)
ax.set_xlabel("Wall Time [s]", fontsize=24)
ax.set_ylabel("$s^*_\mathrm{best}$", fontsize=24)

ax.tick_params(labelsize=24)

plt.scatter(wall_times, bests, s=100)
plt.xscale('log')
plt.savefig("../../thesis_paper/res/wall_time.png", bbox_inches="tight")
```

```python
fig = plt.figure()
ax = fig.gca()

ax.set_title("$s^*_\mathrm{best}$ versus Function Calls", fontsize=24)
ax.set_xlabel("Function Calls", fontsize=24)
ax.set_ylabel("$s^*_\mathrm{best}$", fontsize=24)

ax.tick_params(labelsize=24)

plt.scatter(fn_calls, bests, s=100)
plt.xscale('log')
plt.savefig("../../thesis_paper/res/fn_calls.png", bbox_inches="tight")
```
